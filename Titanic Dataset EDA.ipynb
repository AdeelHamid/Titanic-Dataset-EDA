# Dealing the null values

```python
# as we see there exist null values in the data NaN (Not a number). 
# First we will check how much values are null
df.isnull().sum()  
```

* As we can see there are 106 null values in the age column and we can see it is important data so we will not drop this column. We will impute this
data by evaluating the values
* There are 582 null values in the deck column. We can not impute this as is is more than 50%. So we will try to drop this column.
* There are 2 null values in the embark town and embarked. We will try to impute or if the data of same row is missing more than 2 cells. we will drop rows

```python
# We can confirm missing values by different graphs.
# Create a heatmap of the missing values
plt.figure(figsize=(8, 6))
sns.heatmap(df.isnull(), cmap='viridis', annot=True)
# Yellow shows the missing values
```

```python
# We can also show bar chart to display missing values
# Create a bar chart of the missing values
plt.bar(df.columns, df.isnull().sum())
plt.xlabel('Column')
# Rotate the column names by 90 degrees
plt.xticks(rotation=90)
plt.ylabel('Number of missing values')
plt.title('Missing values in the Titanic data set')
plt.show()
```

```python
# Dropping the deck column will be the best so we drop it
df.drop('deck', axis=1, inplace=True)
```

```python
# checking the dataset after dropping deck now the columns left are 14
df.isnull().sum()
```

```python
# imputing the embarked missing values with mode
mode=df['embarked'].mode()[0]
df['embarked'].fillna(mode, inplace=True)
```

```python
# imputing the embark town missing values with mode
mode=df['embark_town'].mode()[0]
df['embark_town'].fillna(mode, inplace=True)
```

```python
df.isnull().sum()
```

```python
# now we evaluate the age column. first sorting the data with respect to age
df1=df.sort_values(by='age', ascending=True)
df1
```

```python
# It may be any dependency of age on other attributes. So, we need to check correlation. first we check correlation matrix then heatmap
numeric_columns = df1.select_dtypes(include=['int64', 'float64'])
correlation_matrix = numeric_columns.corr()
print(correlation_matrix)
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()

```

we can see there is not any strong relationship of age with other attributes. So, we will follow other rules for imputation of age.

We will try to see if there any outlier in the age column which can make a drastic change in the mean of age

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=df1['age'], orient='h')
plt.title("Boxplot of Age")
plt.show()
```

```python
import matplotlib.pyplot as plt
# Create a box plot with different colors for male and female
plt.figure(figsize=(8, 6))
sns.boxplot(x="sex", y="age", data=df1, palette={"male": "blue", "female": "pink"})
plt.title("Box Plot of Age by Gender")
plt.show()
```

```python

# Filter the dataset to include only male passengers
male_data = df1[df1["sex"] == "male"]

# Create a box plot for male passengers
plt.figure(figsize=(8, 6))
sns.boxplot(x="sex", y="age", data=male_data, palette={"male": "blue"})
plt.title("Box Plot of Age for Male Passengers")

# Calculate the upper bound of the IQR for male passengers
q3 = male_data['age'].quantile(0.75)
iqr = q3 - male_data['age'].quantile(0.25)
upper_bound = q3 + 1.5 * iqr
print(f"Upper bound of IQR for male passengers: {upper_bound}")

plt.show()
```

```python

# Filter the dataset to include only male passengers
male_data = df1[df1['sex'] == 'male']

# Sort the data by age in ascending order
male_data_sorted = male_data.sort_values('age')

# Find the first row where age exceeds 68.5
first_age_above_68_5 = male_data_sorted['age'].searchsorted(68.5, side='right')

# Calculate the mean age of male passengers up to the point where age is 68.5 or less
mean_age_until_68_5 = male_data_sorted['age'][:first_age_above_68_5].mean()

print(f"Mean age of male passengers until age does not exceed 68.5: {mean_age_until_68_5:.2f}")
```

```python
# Filter the dataset to include only male passengers
male_data = df1[df['sex'] == 'male']

# Filter the male data until age does not exceed 68.5
male_data_filtered = male_data[male_data['age'] <= 68.5]

# Create a box plot for the filtered age data
plt.figure(figsize=(8, 6))
plt.boxplot(male_data_filtered['age'], vert=False)
plt.title("Box Plot of Age for Male Passengers (age <= 68.5)")
plt.xlabel("age")
plt.show()
```

```python
# Calculate the mean age of male passengers until age does not exceed 68.5
male_data = df1[df1['sex'] == 'male']
mean_age_until_68_5 = male_data[male_data['age'] <= 68.5]['age'].mean()

# Impute the mean age for null values in the "Age" column for male passengers
df1.loc[(df1['sex'] == 'male') & (df1['age'].isnull()), 'age'] = mean_age_until_68_5

# Verify that null values have been imputed for males only
print(df1['age'].isnull().sum())  # Should output the count of null values
```

```python
df1.isnull().sum() # These are left only 37 values which is for female. So we will impute these for the mean of females
```

```python
female_data = df1[df1['sex'] == 'female']
mean_age_female = female_data['age'].mean()

# Impute the mean age for null values in the "Age" column for females
df1.loc[(df1['sex'] == 'female') & (df1['age'].isnull()), 'age'] = mean_age_female

# Verify that null values have been imputed for females
print(df1['age'].isnull().sum())  # Should output the count of null values
```

# Data Visualization

```python

# Calculate the number of survivors by sex
survivors_by_sex = df1.groupby(['sex', 'survived']).size().unstack(fill_value=0)

# Display the counts in numerical form
print("Number of survivors by sex:")
print(survivors_by_sex)

# Create a bar graph with different colors for survived and sex
plt.figure(figsize=(8, 6))
sns.countplot(x="sex", hue="survived", data=df1, palette={0: "red", 1: "blue"})
plt.title("Survivors by Sex")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.show()
```

```python
# Calculate the number of survivors by gender and Pclass
survivors_by_gender_pclass = df1.groupby(['sex', 'class', 'survived']).size().unstack(fill_value=0)

# Display the counts in numerical form
print("Number of survivors by gender and Pclass:")
print(survivors_by_gender_pclass)

# Create a bar graph with different colors for survived and Pclass
plt.figure(figsize=(12, 8))
sns.countplot(x="sex", hue="pclass", data=df1, palette="Set2", hue_order=[1, 2, 3])
plt.title("Survivors by Gender and Pclass")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.legend(title="Pclass", labels=["1st", "2nd", "3rd"])
plt.show()
```

```python

# Calculate the number of survivors and non-survivors by Pclass
survivors_by_pclass = df1.groupby(['pclass', 'survived']).size().unstack(fill_value=0)

# Display the counts in numerical form
print("Number of survivors and non-survivors by Pclass:")
print(survivors_by_pclass)

# Create a bar graph with different colors for survived and non-survived
plt.figure(figsize=(12, 8))
sns.countplot(x="pclass", hue="survived", data=df1, palette={0: "red", 1: "blue"}, hue_order=[0, 1])
plt.title("Survivors and Non-Survivors by Pclass")
plt.xlabel("Pclass")
plt.ylabel("Count")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.show()
```

```python
# Calculate the number of survivors and non-survivors by Pclass and Sex
survivors_by_pclass_sex = df1.groupby(['pclass', 'sex', 'survived']).size().unstack(fill_value=0)

# Display the counts in numerical form
print("Number of survivors and non-survivors by Pclass and Sex:")
print(survivors_by_pclass_sex)

# Create a bar graph with different colors for survived and non-survived
plt.figure(figsize=(12, 8))
sns.countplot(x="pclass", hue="survived", data=df1, palette={0: "red", 1: "blue"}, hue_order=[0, 1])
plt.title("Survivors and Non-Survivors by Pclass")
plt.xlabel("Pclass")
plt.ylabel("Count")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.show()
```
